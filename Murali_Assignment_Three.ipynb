```
import pandas as pd
import nltk
import re
import string

df = pd.read_csv("my_cse.csv")

#Removing special characters
df["cln_abstract"] = df["Abstract"].str.replace(r"[^a-zA-Z ]+", " ").str.strip()

#Punctuation removal
df['cln_abstract'] = df['cln_abstract'].str.replace('[^\w\s]','')

#remove numbers
df['cln_abstract'] = df['cln_abstract'].str.rstrip(string.digits)

#Stopwords removal
stop = stopwords.words('english')
df['cln_abstract'] = df['cln_abstract'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

#Lower casing
df['cln_abstract'] = df['cln_abstract'].apply(lambda x: " ".join(x.lower() for x in x.split()))


#Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
df['cln_abstract'][:5].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))


#Lemmatization
df['cln_abstract'] = df['cln_abstract'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```

```
# Write your code here
tf_result = (df['cln_abstract']).apply(lambda x: pd.value_counts(x.split(" "))).sum(axis = 0).reset_index()
tf_result.columns = ['cln_abstract','tf']
tf_result




```

```
import numpy as np
#calculating tf*idf
for i,j in enumerate(tf_result['cln_abstract']):
  tf_result.loc[i, 'idf'] = np.log(df.shape[0]/(len(df[df['cln_abstract'].str.contains(j)])))
tf_result['tf*idf'] = tf_result['tf'] * tf_result['idf']
tf_result
```

```
#2.2
#importing required libraries
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from nltk.corpus import stopwords
import numpy as np
import numpy.linalg as LA
```

```
tn_set = df['cln_abstract'].values.tolist() #train set
tt_set = "bwill see realdonaldtrump family getting vaccinated covid model behavior covidvaccine" #Query taken from one of the tweets which is called test set
tt_set = [tt_set]
stopWords = stopwords.words('english')
vzer = CountVectorizer(stop_words = stopWords)
#print(vectorizer)
transformer = TfidfTransformer()
#print(transformer)
```

```
tnVectorizerArray = vzer.fit_transform(tn_set).toarray()
ttVectorizerArray = vzer.transform(tt_set).toarray()
cx = lambda a, b : np.inner(a, b)/(LA.norm(a)*LA.norm(b))
cosine_values = []
```

```
for vector in tnVectorizerArray:
        for testV in ttVectorizerArray:
            cosine = cx(vector, testV)
            cosine_values.append(cosine)
```

```
df_result = df.filter(['cln_abstract'], axis=1)
se = pd.Series(cosine_values)
df_result['Cosine_similarity'] = se.values
df_result
```
